```python
class Node :
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None

class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval
            

    def InsertMiddle(self, middlenode, newdata):
        if middlenode is None:
            print("The mentioned node is absent")
            return
        NewNode = Node(newdata)
        NewNode.nextval = middlenode.nextval
        middlenode.nextval = NewNode
        NewNode.prevval = middlenode
        if NewNode.nextval:
            NewNode.nextval.prevval = NewNode

    def DeleteByKeyword(self, keyword):
        current = self.headval
        while current:
            if current.dataval == keyword:
                if current.prevval:
                    current.prevval.nextval = current.nextval
                if current.nextval:
                    current.nextval.prevval = current.prevval
                del current
                return
            current = current.nextval

Node1 = Node("Mon")
Node2 = Node("Tue")
Node3 = Node("Thu")

dLink = DoubleLinkedList()
dLink.headval = Node1
Node1.nextval = Node2
Node2.prevval = Node1
Node2.nextval = Node3
Node3.prevval = Node2

print("Double Linked List awal :")
dLink.listprint()

dLink.InsertMiddle(Node2, "Wed")

print("\nDouble Linked List setelah penyisipan tengah :")
dLink.listprint()

dLink.DeleteByKeyword("Wed")
print("\nDouble Linked List setelah penghapusan :")
dLink.listprint()

```

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class CircularLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        if not self.headval:
            return
        printval = self.headval
        while True:
            print(printval.dataval)
            printval = printval.nextval
            if printval == self.headval:
                break

    def InsertEnd(self, newdata):
        NewNode = Node(newdata)
        if not self.headval:
            self.headval = NewNode
            NewNode.nextval = self.headval
        else:
            temp = self.headval
            while temp.nextval != self.headval:
                temp = temp.nextval
            temp.nextval = NewNode
            NewNode.nextval = self.headval

    def DeleteEnd(self):
        if not self.headval:
            print("Circular Linked List is empty")
            return
        if self.headval.nextval == self.headval:
            self.headval = None
        else:
            temp = self.headval
            while temp.nextval.nextval != self.headval:
                temp = temp.nextval
            temp.nextval = self.headval

node1 = Node("Mon")
node2 = Node("Tue")
node3 = Node("Wed")

cLink = CircularLinkedList()
cLink.headval = node1
node1.nextval = node2
node2.nextval = node3
node3.nextval = cLink.headval

print("Circular Linked List awal :")
cLink.listprint()

cLink.InsertEnd("Thu")
print("\nCircular Linked List setelah penyisipan akhir :")
cLink.listprint()

cLink.DeleteEnd()
print("\nCircular Linked List setelah penghapusan akhir :")
cLink.listprint()
```
