```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None

class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    def InsertMiddle(self, middle_node, newdata):
        if middle_node is None:
            print("The mentioned node is absent")
            return
        NewNode = Node(newdata)
        NewNode.nextval = middle_node.nextval
        middle_node.nextval = NewNode
        NewNode.prevval = middle_node
        if NewNode.nextval:
            NewNode.nextval.prevval = NewNode

    def DeleteByKeyword(self, keyword):
        current = self.headval
        while current:
            if current.dataval == keyword:
                if current.prevval:
                    current.prevval.nextval = current.nextval
                if current.nextval:
                    current.nextval.prevval = current.prevval
                del current
                return
            current = current.nextval

# Buat 3 node untuk Double Linked List
node1 = Node("Node 1")
node2 = Node("Node 2")
node3 = Node("Node 3")

# Inisialisasi Double Linked List
dLinkedList = DoubleLinkedList()
dLinkedList.headval = node1
node1.nextval = node2
node2.prevval = node1
node2.nextval = node3
node3.prevval = node2

# Tampilkan Double Linked List awal
print("Double Linked List awal:")
dLinkedList.listprint()

# Sisipkan node baru di tengah (setelah node kedua)
dLinkedList.InsertMiddle(node2, "Node Baru")

# Tampilkan Double Linked List setelah penyisipan tengah
print("\nDouble Linked List setelah penyisipan tengah:")
dLinkedList.listprint()

# Hapus node dengan kata kunci "!"
dLinkedList.DeleteByKeyword("Node 3")

# Tampilkan Double Linked List setelah penghapusan dengan kata kunci "!"
print("\nDouble Linked List setelah penghapusan dengan kata kunci '!':")
dLinkedList.listprint()
```

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class CircularLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        if self.headval is None:
            print("List is empty")
            return
        printval = self.headval
        while True:
            print(printval.dataval)
            printval = printval.nextval
            if printval == self.headval:
                break

    def InsertEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
        else:
            last = self.headval
            while last.nextval != self.headval:
                last = last.nextval
            last.nextval = NewNode
            NewNode.nextval = self.headval

    def DeleteEnd(self):
        if self.headval is None:
            print("List is empty")
            return
        elif self.headval.nextval == self.headval:
            self.headval = None
            return
        else:
            last = self.headval
            while last.nextval.nextval != self.headval:
                last = last.nextval
            last.nextval = self.headval

# Buat 3 node untuk Circular Linked List
node1 = Node("Node 1")
node2 = Node("Node 2")
node3 = Node("Node 3")

# Inisialisasi Circular Linked List
cLinkedList = CircularLinkedList()
cLinkedList.headval = node1
node1.nextval = node2
node2.nextval = node3
node3.nextval = node1  # Menghubungkan kembali ke node pertama untuk membuatnya menjadi circular

# Tampilkan Circular Linked List awal
print("Circular Linked List awal:")
cLinkedList.listprint()

# Sisipkan node baru di akhir
cLinkedList.InsertEnd("Node Baru")

# Tampilkan Circular Linked List setelah penyisipan akhir
print("\nCircular Linked List setelah penyisipan akhir:")
cLinkedList.listprint()

# Hapus node terakhir
cLinkedList.DeleteEnd()

# Tampilkan Circular Linked List setelah penghapusan akhir
print("\nCircular Linked List setelah penghapusan akhir:")
cLinkedList.listprint()
```
